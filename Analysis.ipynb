### Create a model using these variables

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.preprocessing import scale
from sklearn.cross_validation import train_test_split
```

```python
raw_df = raw_df.fillna(value='No Value', downcast=raw_df.dtypes.to_dict())
```

On impose un type String pour les object sinon DictVectorizer n'est pas content

```python
for c in raw_df.columns[raw_df.dtypes != 'object']:
    raw_df[c] = scale(raw_df[c].astype(np.float))
for c in raw_df.columns[raw_df.dtypes == 'object']:
    raw_df[c] = raw_df[c].astype(str)
```

```python
raw_dict = raw_df.drop('target',1).T.to_dict().values()
```

```python
vec = DictVectorizer()
X = vec.fit_transform(raw_dict)
```

On veux avoir un X relativement normalisé afin d'avoir des bonnes performances pour le SVM qui est particulierement sensible.

```python
X.mean()
```

```python
y = np.array(raw_df.target.astype(np.int))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

### Tree-based feature selection

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import confusion_matrix, f1_score
```

```python
ada_clf = AdaBoostClassifier()
ada_clf.fit(X, y)
```

```python
f1_score(y_test, ada_clf.predict(X_test))
```

```python
feature_importance = pd.DataFrame(np.column_stack((np.array(vec.feature_names_), ada_clf.feature_importances_)), columns=['feature', 'importance'])
feature_importance.importance = feature_importance.importance.astype(np.float)
feature_importance.dtypes
```

```python
feature_importance[feature_importance.importance > 0].sort(columns=['importance'], ascending=False)
```

### One shot linearSVC

```python
from sklearn import svm
```

```python
clf = svm.LinearSVC(verbose=True)
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
print clf.predict(X_test).sum()
print y_test.sum()
print f1_score(y_test, clf.predict(X_test))
```

### One shot SVC rbf

```python
clf = svm.SVC(cache_size=800)
clf.fit(X_train, y_train)
```

```python
y_predicted = clf.predict(X_test)
```
